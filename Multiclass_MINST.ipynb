```python
#2.2 Choosing the best hyper-parameter
```

```python
from sklearn.multiclass import OneVsRestClassifier
def runLRmodel_2(trials, data, alpha):     
    model_acc_with_train=0, 
    model_acc_with_validation=0, 
    model_coef = 0 
    for i in range(0,trials):
        Dtrain, Dtest = train_test_split(data, test_size=0.3)       
        classifier = OneVsRestClassifier(LogisticRegression(penalty='l1', C=1.0/alpha))
        classifier.fit(Dtrain[:,1:], Dtrain[:,0])             
        y_predict_with_train = classifier.predict(Dtrain[:,1:])      
        model_acc_with_train += accuracy_score(Dtrain[:,0], y_predict_with_train)                 
        y_predict_with_validation = classifier.predict(Dtest[:,1:])   
        model_acc_with_validation += accuracy_score(Dtest[:,0], y_predict_with_validation)
         
    model_acc_with_train /= trials    
    model_acc_with_validation /= trials    
       
    return np.round(model_acc_with_train, decimals=2), np.round(model_acc_with_validation, decimals=2)
```

```python
trial_num = 10
alpha_val = [0.1, 1, 3, 10, 33, 100, 333, 1000, 3333, 10000, 33333]
acc_train = np.zeros(len(alpha_val))
acc_validation = np.zeros(len(alpha_val)) 
index = 0 
for l in alpha_val: 
    acc_train[index], acc_validation[index] = runLRmodel_2(trial_num, mnistdata, np.float(l))    
    index += 1
```

```python
#Evaluate our model
print ('MNIST Model Accuracy with Train is:      {}'.format(acc_train)) 
print ('MNIST Model Accuracy with validation is: {}'.format(acc_validation))
```

```python
maxindextrain  = np.argmax(acc_train)
bestalphatrain = alpha_val[maxindextrain]
print ('Best alpha training: {}'.format(bestalphatrain))
maxindexvalidation  = np.argmax(acc_validation)
best_alpha_validation = alpha_val[maxindexvalidation]
print ('Validation Data Best Alpha: {}'.format(best_alpha_validation))
```

```python
# Plotting the accuracy curve
plt.plot(range(0,len(alpha_val)),  acc_train, color='r', label='Train')
plt.plot(range(0,len(alpha_val)), acc_validation, color='b', label='Validation')
#Replacing lables of x and y axes
plt.xticks(range(0,len(alpha_val)), alpha_val, rotation='vertical') 

#Best value of alpha and lambda 
plt.plot((maxindextrain, maxindextrain), (0, acc_train[maxindextrain]), ls='dotted', color='r')
plt.plot((maxindexvalidation, maxindexvalidation), (0, acc_validation[maxindexvalidation]), ls='dotted', color='b') 
# Setting y-axis from 0 to 1.0
axes = plt.gca() 

#Setting the y-axis from 0 to 1.0
axes.set_ylim([0, 1.0])

plt.legend(loc="lower left")
```

```python
#2.2(2)- Evaluating the performance of prediction of test data
```

```python

#trial number to be set as 10 
trial_num = 10
model_acc=0
model_precision=0
model_recall=0
model_coef = 0 
# Using best_alpha_validation run again tp obtain coeffcient 
   
Dtrain, Dtest = train_test_split(mnistdata, test_size=0.3)  
classifier = OneVsRestClassifier(LogisticRegression(penalty='l1', C=1.0/best_alpha_validation))    
classifier.fit(Dtrain[:,1:], Dtrain[:,0])               
    
y_predict_validation= classifier.predict(Dtest[:,1:])   
model_coef += classifier.coef_     
model_acc += accuracy_score(Dtest[:,0], y_predict_validation)   
model_precision += precision_score(Dtest[:,0], y_predict_validation, average='weighted')    
model_recall += recall_score(Dtest[:,0], y_predict_validation, average='weighted')         

print ('Total number of non-zero features in the final model ', np.count_nonzero(model_coef[0]))
```
